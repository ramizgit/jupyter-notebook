```python
import matplotlib.pyplot as plt
import numpy as np

plt.plot([10, 25,20])
plt.grid(True)
plt.show()
```

```python
plt.plot([1,2],[1,2])
plt.grid(True)
plt.show()
```

```python
plt.plot([-3, -2, 5, 0], [1, 6, 4, 3], "g--")
plt.axis([-10,10,0,10])
plt.grid(True)
plt.show()
```

```python
x = np.linspace(-2, 2, 500)
y = x**2

plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(-2, 2, 500)
y = x**2

plt.plot(x, y, "r--")
plt.title("Square function")
plt.xlabel("x")
plt.ylabel("y = x**2")
plt.grid(True)
plt.show()
```
