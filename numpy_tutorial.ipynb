```python
g = np.arange(10)
print(g)
print("Rank:", g.ndim)
print("Shape:", g.shape)
#change shape
g.shape = (2, 5)
print(g)
print("Rank:", g.ndim)
print("Shape:", g.shape)
#reshape via new array
g2 = g.reshape(2,5)
print(g2)
print("Rank:", g2.ndim)
print("Shape:", g2.shape)

#chagne element at a particular index
g2[0,0]=99999
print(g2)
```

```python
###ravel
print("ravel")

ravelarr = g2.ravel()

#change 
ravelarr[0] = 33333

print(ravelarr)
print(g2)
```

```python
###flatten
flatarr=g2.flatten()
print(flatarr)
flatarr[0]=88888
print(flatarr)
print(g2)
```

```python
#stacking
q1 = np.full((3,4), 1.0)
q2 = np.full((4,4), 2.0)
q3 = np.full((3,4), 3.0)

#horizontal
q4 = np.vstack((q1, q2, q3))
print(q4)

#vertical
q5 = np.hstack((q1, q3))
print(q5)
```

```python
#split
a = np.arange(20).reshape(4,5)
a

#vertical split
v1,v2,v3,v4 = np.vsplit(a, 4)
print(v1)
print(v2)
print(v3)
print(v4)
print("================")
#horizontal split
a = np.arange(24).reshape(6,4)
print(a)
v5,v6 = np.hsplit(a,2)
print(v5)
print(v6)

```

```python
#vectorization
x_coords = np.arange(0, 1024)  # [0, 1, 2, ..., 1023]
y_coords = np.arange(0, 768)   # [0, 1, 2, ..., 767]
X, Y = np.meshgrid(x_coords, y_coords)
print(X.shape)
print(Y.shape)

data = np.sin(X * Y / 40.5)
print(data)
```
